```python
import matplotlib.pyplot as plt
import numpy as np

# Set Matplotlib Parameters
plt.style.use('bmh')
plt.rc('text', usetex=True)   #  Might require installing TeX fonts
plt.rc('axes', facecolor='w', labelcolor='b', edgecolor='k', labelsize=18)
plt.rcParams['font.size'] = 16
plt.rcParams['legend.fontsize'] = 16
plt.rcParams['xtick.labelsize'] = 16
plt.rcParams['ytick.labelsize'] = 18
```

```python
# short: 5430.36
# short: Count: 45808
# medium: 6609.75
# medium: Count: 45808
# long: 8044.48
# long: Count: 54204
# None: 15628.60
# None: Count: 67


data_lens = [
    ("Short", 5430),
    ("Medium", 6609),
    ("Long", 8044),
]



plt.figure(figsize=(10, 6))
plt.bar(
    [bucket[0] for bucket in data_lens],
    [bucket[1] for bucket in data_lens],
    color=['#1f77b4', '#ff7f0e', '#2ca02c'],
    alpha=0.7,
    width=0.5
)

plt.ylabel('Average Solution Length (\# of tokens)', color='gray')
plt.xlabel('Solution Length Bucket', color='gray')

plt.savefig("../outputs/sft_bucket_data_len.png")
```
