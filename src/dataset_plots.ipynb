```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from datasets import load_from_disk
```

```python
# Load dataset (assuming it's already loaded as `dataset`)
dataset = load_from_disk("../data/open-r1/OpenR1-Math-220k_filtered")
```

```python
# Ensure short_tokens always stores the shorter generation
short_tokens = dataset["train"]["min_length"]
long_tokens = dataset["train"]["max_length"]

# Compute the token difference
token_diff = np.array(long_tokens) - np.array(short_tokens)
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Create figure and subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
colors = sns.color_palette("tab10")

# Plot histograms
sns.histplot(short_tokens, bins=50, kde=True, color=colors[0], ax=axes[0])
axes[0].set_title("Token Count (Short Generation)")
axes[0].set_xlabel("Token Count")
axes[0].set_ylabel("Frequency")

sns.histplot(long_tokens, bins=50, kde=True, color=colors[1], ax=axes[1])
axes[1].set_title("Token Count (Long Generation)")
axes[1].set_xlabel("Token Count")
axes[1].set_ylabel("Frequency")

sns.histplot(token_diff, bins=50, kde=True, color=colors[2], ax=axes[2])
axes[2].set_title("Token Count Difference")
axes[2].set_xlabel("Token Difference (Long - Short)")
axes[2].set_ylabel("Frequency")

# Show plots
plt.tight_layout()
plt.show()
```
