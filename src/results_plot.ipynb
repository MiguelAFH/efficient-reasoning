```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Data
models = ['deepseek', 'deepseek-dpo']
accuracy = [0.167, 0.167]
token_length = [3727, 3702]

# Create a DataFrame
df = pd.DataFrame({'Model': models, 'Accuracy': accuracy, 'Token Length': token_length})

# Set style and color palette
sns.set_style("whitegrid")
palette = sns.color_palette("tab10")

# Create the bar plot
plt.figure(figsize=(6, 4))
ax = sns.barplot(x='Model', y='Accuracy', data=df, palette=palette)

# Add text labels for token length
for i, (acc, token) in enumerate(zip(accuracy, token_length)):
    plt.text(i, acc + 0.01, f'Token: {token}', ha='center', fontsize=10)

# Labels and title
plt.ylabel("Accuracy")
plt.xlabel("Model")
plt.title("LLM as a judge Accuracy")
plt.ylim(0, 0.3)  # Set y-axis limit

# Show plot
plt.show()
```
